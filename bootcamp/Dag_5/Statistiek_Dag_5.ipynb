# Opdracht 1 - Tevredenheid voorspellen

#### T-test analyse 

```python
# import the csv file
airline = pd.read_csv('/Users/jonathanricardo/Library/CloudStorage/OneDrive-HvA/Semester 4 - AI/Logboek AAI/minor-logboek-aai-2023-jonathan/bootcamp/Dag_3/Airlinesatisfaction.csv')

#Random sample of 100
random_subset = airline.sample(n=300)

# Drop unnecessary columns
random_subset = random_subset.drop(['id', 'Gate location'], axis=1)

# Vul de missende data met
mean_delay = random_subset['Arrival Delay in Minutes'].mean()
random_subset['Arrival Delay in Minutes'] = random_subset['Arrival Delay in Minutes'].fillna(mean_delay)

# Rename columns
random_subset.columns = [c.replace(' ', '_') for c in random_subset.columns]

# Replace categoric data with numberic data
random_subset['Gender'].replace({'Male': 0, 'Female': 1},inplace = True)
random_subset['satisfaction'].replace({'neutral or dissatisfied': 0, 'satisfied': 1},inplace =True)
random_subset['Customer_Type'].replace({'disloyal Customer': 0,'Loyal Customer': 1}, inplace = True)
random_subset['Type_of_Travel'].replace({'Personal Travel': 0, 'Business travel': 1}, inplace = True)
random_subset['Class'].replace({'Eco': 0, 'Eco Plus': 1, 'Business': 2} ,inplace=True)

# Print the first 5 rows of the dataframe
random_subset.head()
```

Hier heb ik de data voorbereidt om te gebruiken in een t-test en andere verdere analysen.

```python
# list of features
features = ['Gender', 'Customer_Type', 'Age', 'Type_of_Travel', 'Class', 'Flight_Distance', 'Inflight_wifi_service', 'Departure/Arrival_time_convenient', 'Ease_of_Online_booking', 'Food_and_drink', 'Online_boarding', 'Seat_comfort', 'Inflight_entertainment', 'On-board_service', 'Leg_room_service', 'Baggage_handling', 'Checkin_service', 'Inflight_service', 'Cleanliness', 'Departure_Delay_in_Minutes', 'Arrival_Delay_in_Minutes']
result_features = []

# t-test on all features
for feature in features:
    summary, result = stats.ttest_ind(random_subset[feature], random_subset['satisfaction'])
    print(f"Feature: {feature},Summary: {summary}, P-value: {result}")

    if result < 0.05:
        result_features.append(feature)
```

Hier kan je alle p waarden zien van alle features in onze data set. Waar wij naar toe willen kijken, is een p-value van 0.05 of lager. Dat betekend namelijk dat er significantie is.

Dit zijn: `Gender`, `Type_of_Travel` en `Departure_Delay_in_Minutes`

# Oefening 2
#### ANOVA - verschil leeftijd & klasse

```python
# ANOVA test for age with class
stats.f_oneway(random_subset['Age'][random_subset['Class'] == 0],
               random_subset['Age'][random_subset['Class'] == 1],
               random_subset['Age'][random_subset['Class'] == 2])
```

De uitgevoerde ANOVA-test onderzoekt of er een significant verschil is in de gemiddelde leeftijden van de drie verschillende vliegklassen. De resultaten van de test geven een F-statistiek en een p-waarde.

De F-statistiek is een maat voor hoeveel de gemiddelden van de groepen verschillen. Een hogere F-statistiek suggereert dat de groepsgemiddelden meer van elkaar verschillen dan je zou verwachten op basis van toeval alleen.

De p-waarde is een maat voor de waarschijnlijkheid dat je de waargenomen resultaten (of extremere) zou krijgen als de nulhypothese waar is. In dit geval is de nulhypothese dat er geen verschil is in de gemiddelde leeftijden van de drie vliegklassen.

Omdat de p-waarde kleiner is dan 0.05, is er sterk bewijs om de nulhypothese te verwerpen. Dit betekent dat er een statistisch significant verschil is in de gemiddelde leeftijden van de drie vliegklassen.

> bij nader inzien zag ik dat alle waarden extreem laag waren. Wellicht betekend dit dat er iets niet helemaal goed is gegaan.

#### ANOVA - verschil flightdistance en klassa

```python
# ANOVA test for flightdistance with class
stats.f_oneway(random_subset['Flight_Distance'][random_subset['Class'] == 0],
               random_subset['Flight_Distance'][random_subset['Class'] == 1],
               random_subset['Flight_Distance'][random_subset['Class'] == 2])
```

De F-statistiek in de ANOVA-test is 15.7642 en de p-waarde is 1.1802452611883205e-06.

De p-waarde is veel kleiner dan 0.05, wat meestal wordt gebruikt als drempel voor statistische significantie. Dit betekent dat we het nulhypothese (dat er geen verschil is in de gemiddelde vluchtafstand tussen de verschillende klassen) verwerpen.

Dus, op basis van deze resultaten, is er een statistisch significant verschil in vluchtafstand tussen de drie vliegklassen.

```python
# Selected features
features = ['Online_boarding', 'Inflight_entertainment', 'On-board_service', 'Leg_room_service', 'Cleanliness', 'Departure_Delay_in_Minutes', 'Class', 'Gender', 'Type_of_Travel']

# Split the data into a training and test set
train_X, test_X, train_y, test_y = train_test_split(random_subset[features], random_subset['satisfaction'], test_size=0.2, random_state=0)

# Initialize the models
models = [LogisticRegression(max_iter=1000), KNeighborsClassifier(n_neighbors=3), svm.SVC(kernel='linear'), svm.SVC(kernel='rbf'), DecisionTreeClassifier()]

# Train the models
percentage = []
for model in models:
    model.fit(train_X, train_y)
    predictions = model.predict(test_X)
    accuracy = accuracy_score(test_y, predictions)
    percentage.append(accuracy)
accuracy_df = pd.DataFrame(percentage, index=['Logistic Regression', 'KNN', 'SVM Linear', 'SVM RBF', 'Decision Tree'], columns=['Accuracy'])
accuracy_df
```
