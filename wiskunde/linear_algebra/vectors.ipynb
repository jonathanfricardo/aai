# Vectors

Een vector is een wiskundig object dat zowel een grootte (of lengte) als een richting heeft. Vectors worden vaak gebruikt om fysieke grootheden te beschrijven die zowel een magnitude als een richting hebben, zoals snelheid of kracht.

Een vector kan worden weergegeven als een geordend paar (of een geordende lijst in hogere dimensies), bijvoorbeeld (3,1). Dit kan ook worden weergegeven als een kolomvector:

$$  
\begin{pmatrix}
3 \\ 
1
\end{pmatrix}
$$
of een rijvector: 
$$
(3, 1)
$$

Enkele eigenschappen van een vector zijn:

- Grootte: De lengte van de vector, vaak berekend met behulp van de Euclidische norm (de vierkantswortel van de som van de kwadraten van de componenten).
- Richting: De richting waarin de vector wijst.
- Rijvector: Een vector weergegeven als een rij in een matrix.
- Kolomvector: Een vector weergegeven als een kolom in een matrix.

Optellen van vectoren: Je telt vectoren bij elkaar op door de corresponderende elementen bij elkaar op te tellen. Als je bijvoorbeeld de vectoren 
$$v = (v1, v2) \\
 w = (w1, w2)$$ 
hebt, dan is het:
$$v + w = (v1 + w1, v2 + w2).$$

Vermenigvuldigen van vectoren: Er zijn twee manieren om vectoren met elkaar te vermenigvuldigen: het inwendig product (ook wel dot product of scalair product genoemd) en het uitwendig product (ook wel cross product of vector product genoemd).

Het inwendig product van twee vectoren 
$$v = (v1, v2) \\ 
 w = (w1, w2)
$$ 

is een getal (scalair) en wordt berekend als 

$$
v \cdot w = v1*w1 + v2*w2.
$$

Dit zijn eigenschappen van het inproduct (dot-product) van twee vectoren en de hoek tussen hen. Ze komen voort uit de formule voor het inproduct in termen van de hoek: V·W = |V||W|cos(α), waarbij α de hoek tussen de vectoren is.

a. Als V·W > 0, dan is de hoek tussen de vectoren kleiner dan 90 graden. Dit komt omdat de cosinus van een hoek tussen 0 en 90 graden positief is, dus als het inproduct positief is, moet de hoek tussen de vectoren kleiner zijn dan 90 graden.

b. Als V·W = 0, dan is de hoek tussen de vectoren 90 graden. Dit komt omdat de cosinus van 90 graden 0 is, dus als het inproduct 0 is, moeten de vectoren loodrecht op elkaar staan.

c. Als V·W < 0, dan is de hoek tussen de vectoren groter dan 90 graden. Dit komt omdat de cosinus van een hoek tussen 90 en 180 graden negatief is, dus als het inproduct negatief is, moet de hoek tussen de vectoren groter zijn dan 90 graden.
