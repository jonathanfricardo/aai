De Manhattan- en Euclidische afstanden zijn twee verschillende manieren om de afstand tussen twee punten in een ruimte te meten.

#### Manhattan Afstand: 
Ook bekend als de taxicab of city block afstand. De Manhattan afstand tussen twee punten is de som van de absolute verschillen van hun coördinaten. Stel, we hebben twee punten a en b met coördinaten (ax, ay) en (bx, by) respectievelijk. De Manhattan afstand tussen deze punten wordt berekend als:

$$
d_{M}(a,b) = |a_{x} - b_{x}| + |a_{y} - b_{y}|
$$

Dit is alsof je door een raster van straten in een stad navigeert (vandaar de naam 'city block'), waarbij je alleen horizontaal en verticaal kunt bewegen, niet diagonaal.


#### Euclidische Afstand:
 Dit is de meest gebruikelijke manier om de afstand tussen twee punten te meten. Het is de "rechte lijn" afstand tussen de punten. In een tweedimensionale ruimte wordt de Euclidische afstand berekend als de vierkantswortel van de som van de kwadraten van de verschillen tussen de coördinaten van de punten. Voor de punten a en b zoals hierboven gedefinieerd, wordt de Euclidische afstand berekend als:

$$
d_{E}(a,b) = \sqrt{(a_{x} - b_{x})^{2} + (a_{y} - b_{y})^{2}}
$$

Deze afstand is gebaseerd op de stelling van Pythagoras en is de meest directe route tussen twee punten.



```python
import matplotlib.pyplot as plt
import numpy as np

# Coördinaten van de punten
ax, ay = (1, 3)
bx, by = (4, 6)

# Bereken de punten langs de Manhattan-afstand
mx = [ax, bx, bx]
my = [ay, ay, by]

# Bereken de punten langs de Euclidische afstand
ex = np.linspace(ax, bx, 100)
ey = np.linspace(ay, by, 100)

# Plot de punten en de afstanden
plt.plot([ax, bx], [ay, by], 'ro')  # De punten
plt.plot(mx, my, label='Manhattan distance')  # De Manhattan-afstand
plt.plot(ex, ey, label='Euclidean distance')  # De Euclidische afstand

# Voeg een legenda toe
plt.legend()

# Grid toevogen
plt.grid()
plt.xlim(0, 7)
plt.ylim(0, 7)

# Toon de plot
plt.show()
```
