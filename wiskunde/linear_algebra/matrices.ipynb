# Matrices

Een matrix is een tweedimensionale gegevensstructuur die bestaat uit rijen en kolommen. Elke cel in de matrix bevat een waarde. In de context van programmeren kunnen matrices worden gebruikt om meerdere gegevens van hetzelfde type op te slaan.

Matrices zijn bijzonder belangrijk in AI om verschillende redenen:

Gegevensrepresentatie: In machine learning en AI worden gegevens vaak weergegeven als matrices. Bijvoorbeeld, een afbeelding kan worden weergegeven als een matrix van pixelwaarden.

Efficiënte berekeningen: Veel algoritmen in AI, zoals neurale netwerken, zijn gebaseerd op matrixberekeningen. Deze berekeningen kunnen efficiënt worden uitgevoerd met matrices.

Lineaire algebra: Matrices zijn een fundamenteel onderdeel van de lineaire algebra, een tak van wiskunde die veel wordt gebruikt in AI. Met matrices kunnen complexe lineaire vergelijkingen worden opgelost die vaak voorkomen in AI-toepassingen.


### Optellen & Aftrekken

Het optellen en aftrekken van matrices is vrij eenvoudig en volgt een element-voor-element methode. Dit betekent dat je gewoon de overeenkomstige elementen in de twee matrices bij elkaar optelt of van elkaar aftrekt.

Hier zijn de stappen om twee matrices op te tellen of af te trekken:

Zorg ervoor dat de twee matrices dezelfde dimensies hebben. Je kunt alleen matrices optellen of aftrekken als ze hetzelfde aantal rijen en kolommen hebben.

Voeg de overeenkomstige elementen samen. Dit betekent dat je het element in de eerste rij en eerste kolom van de eerste matrix optelt bij het element in de eerste rij en eerste kolom van de tweede matrix, enzovoort voor alle elementen.

Optellen:
$$  
\begin{pmatrix}
3 & 6 & 1 \\ 
1 & 2 & 8 \\
5 & 6 & 9
\end{pmatrix}
+
\begin{pmatrix}
9 & 2 & 5 \\ 
6 & 8 & 1 \\
3 & 2 & 5
\end{pmatrix}
=
\begin{pmatrix}
12 & 8 & 6 \\ 
7 & 10 & 9 \\
8 & 8 & 14
\end{pmatrix}
$$

Aftrekken:
$$  
\begin{pmatrix}
9 & 8 & 7 \\ 
6 & 5 & 4 \\
3 & 2 & 1
\end{pmatrix}
-
\begin{pmatrix}
1 & 2 & 3 \\ 
4 & 5 & 6 \\
7 & 8 & 9
\end{pmatrix}
=
\begin{pmatrix}
8 & 6 & 4 \\ 
2 & 0 & -2 \\
-4 & -6 & -8
\end{pmatrix}
$$

### Hadamardproduct

Het Hadamardproduct, ook bekend als het elementsgewijze product, is een wiskundige operatie die twee matrices van dezelfde dimensies neemt en een nieuwe matrix produceert waarvan elk element het product is van de overeenkomstige elementen van de originele matrices.

$$  
\begin{pmatrix}
1 & 2 & 3 \\ 
4 & 5 & 6 \\
7 & 8 & 9
\end{pmatrix}
\odot
\begin{pmatrix}
9 & 8 & 7 \\ 
6 & 5 & 4 \\
3 & 2 & 1
\end{pmatrix}
=
\begin{pmatrix}
9 & 16 & 21 \\ 
24 & 25 & 24 \\
21 & 16 & 9
\end{pmatrix}
$$

### Vermenigvuldiging

Gegeven twee matrices A en B, waarbij A een m x n matrix is en B een n x k matrix, kan de vermenigvuldiging C = A · B worden berekend. De voorwaarde hiervoor is dat het aantal kolommen (n) van de eerste matrix (A) gelijk moet zijn aan het aantal rijen (n) van de tweede matrix (B).

Het resultaat is een nieuwe matrix C met m rijen en k kolommen. Elk element C(m; k) in de resulterende matrix wordt berekend als het dot product van de m-de rij van A en de k-de kolom van B.

Hier is een voorbeeld van hoe dit werkt met twee 2x2 matrices:

Laat $$ A =  \begin{pmatrix} a & b \\ c & d \end{pmatrix} en B = \begin{pmatrix} e & f \\ g & h \end{pmatrix}$$ 

Dan is $$ C = A · B = \begin{pmatrix} a*e + b*g & a*f + b*h \\ c*e + d*g & c*f + d*h \end{pmatrix}$$ 

Elk element in de resulterende matrix C is het resultaat van het vermenigvuldigen van de overeenkomstige rij in A met de overeenkomstige kolom in B.


### Matrix-vector vermenigvuldiging

Matrix-vector vermenigvuldiging werkt op een vergelijkbare manier als matrix-matrix vermenigvuldiging, maar het resultaat is een vector in plaats van een matrix.

Gegeven een m x n matrix A en een n x 1 vector v, kan de vermenigvuldiging c = A · v worden berekend. De voorwaarde hiervoor is dat het aantal kolommen (n) van de matrix (A) gelijk moet zijn aan het aantal rijen (n) van de vector (v).

Het resultaat is een nieuwe vector c met m rijen. Elk element c(m) in de resulterende vector wordt berekend als het dot product van de m-de rij van A en de vector v.

Hier is een voorbeeld van hoe dit werkt met een 3x2 matrix en een 2x1 vector:

Laat $$A = \begin{pmatrix} a_{1,1} & a_{1,2} \\ a_{2,1} & a_{2,2} \\ a_{3,1} & a_{3,2} \end{pmatrix}$$ en $$ v = \begin{pmatrix} v_{1} \\ v_{2} \end{pmatrix}$$

Dan is $$c = A · v = \begin{pmatrix} a_{1,1}*v_{1} + a_{1,2}*v_{2} \\ a_{2,1}*v_{1} + a_{2,2}*v_{2} \\ a_{3,1}*v_{1} + a_{3,2}*v_{2} \end{pmatrix}$$

Elk element in de resulterende vector c is het resultaat van het vermenigvuldigen van de overeenkomstige rij in A met de vector v.

### Inverse matrix bepalen

In veel machine learning-algoritmen worden lineaire vergelijkingen opgelost, zoals die voorkomen in lineaire regressie of logistische regressie. Het gebruik van de inverse van een matrix kan een efficiënte manier zijn om deze vergelijkingen op te lossen. In plaats van herhaaldelijk matrices te moeten inverseren, kan het berekenen van de inverse eenmalig worden gedaan, wat in sommige gevallen tot snellere berekeningen kan leiden.

Het berekenen van de inverse van een matrix is een meerstappenproces. Hier is hoe je het doet voor een 3x3 matrix:

Gegeven een matrix A:

$$ A = \begin{pmatrix} 1 & 2 & 6 \\ 0 & 6 & 2 \\ 0 & 4 & 4 \end{pmatrix}$$

Stap 1) Bepaal de determinant (D) van A. Dit kan worden gedaan met de formule:

$$ D = a(ei−fh)−b(di−fg)+c(dh−eg)$$

Voor matrix A, is dit: 

$$ D = 1*(6*4 - 2*4) - 2*(0 - 2*0) + 6*(0 - 6*0) = 16$$

Stap 2) Bepaal de getransponeerde matrix (A^T) van A. Dit wordt gedaan door de rijen en kolommen van A om te wisselen:

$$ A^T = \begin{pmatrix} 1 & 0 & 0 \\ 2 & 6 & 4 \\ 6 & 2 & 4 \end{pmatrix} $$

Stap 3) Bepaal de determinant van alle 2x2 submatrices van A^T. Dit wordt gedaan door voor elke element in A^T de determinant te berekenen van de 2x2 matrix die wordt gevormd door de rijen en kolommen van A^T te verwijderen die het element kruisen:

$$ M = \begin{pmatrix} 6*4 - 2*4 & 2*4 - 6*4 & 2*2 - 6*6 \\ 0*4 - 2*0 & 1*4 - 6*0 & 1*2 - 6*0 \\ 0*4 - 6*0 & 1*4 - 2*0 & 1*6 - 2*0 \end{pmatrix} = \begin{pmatrix} 16 & -16 & -32 \\ 0 & 4 & 2 \\ 0 & 4 & 6 \end{pmatrix} $$ 

Stap 4) Bepaal de adjunct matrix van A. Dit wordt gedaan door de tekenen van de elementen in de matrix van stap 3 te veranderen volgens het patroon (+ - + - + - ...):

$$ Adj(A) = \begin{pmatrix} 16 & 16 & -32 \\ 0 & 4 & -2 \\ 0 & -4 & 6 \end{pmatrix} $$

Stap 5) Deel het resultaat van stap 4 door de determinant (D) van A. Dit geeft de inverse van A:

$$ A^{-1} = \frac{1}{D} * Adj(A) = \frac{1}{16} * \begin{pmatrix} 16 & 16 & -32 \\ 0 & 4 & -2 \\ 0 & -4 & 6 \end{pmatrix} = \begin{pmatrix} 1 & 1 & -2 \\ 0 & \frac{1}{4} & \frac{-1}{8}  \\ 0 & -\frac{1}{4}  & \frac{3}{8}  \end{pmatrix} $$

Stap 6) Controleer of A · A^−1 = I (de identiteitsmatrix). Als dit waar is, dan is de berekening correct. In dit geval, als je A vermenigvuldigt met A^−1, krijg je inderdaad de identiteitsmatrix, dus de berekening is correct.
